```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('datosAsteroide.csv')
# Convert the close_approach_date to datetime format
data ['close_approach_date'] = pd.to_datetime(data['close_approach_date'])

# Set the plot size for better visibility
# Set the plot size for better visibility
plt.figure(figsize=(15, 22))

# Scatter plot: Relative velocity vs Miss distance
plt.subplot(3, 1, 1)
plt.scatter(data['relative_velocity_km_h'], data['miss_distance_km'], alpha=0.5)
plt.title('Velocidad Relativa vs. Distancia Mínima de Aproximación')
plt.xlabel('Velocidad Relativa (km/h)')
plt.ylabel('Distancia Mínima de Aproximación (km)')
plt.grid(True)


# Histogram: Distribution of estimated diameter max
plt.subplot(3, 1, 2)
plt.hist(data['estimated_diameter_max_km'], bins=30, edgecolor='black')
plt.title('Distribución del Diámetro Máximo Estimado')
plt.xlabel('Diámetro Máximo Estimado (km)')
plt.ylabel('Frecuencia')
plt.grid(True)
# Line plot: Number of close approaches over time
data['year'] = data['close_approach_date'].dt.year
yearly_approaches = data.groupby('year').size()
plt.subplot(3, 1, 3)
plt.plot(yearly_approaches.index, yearly_approaches.values, marker='o')
plt.title('Frecuencia de Aproximaciones Cercanas a lo Largo del Tiempo')
plt.xlabel('Año')
plt.ylabel('Número de Aproximaciones Cercanas')
plt.grid(True)
# Show plot with all conclusions
plt.figure(figsize=(15, 18))
plt.subplot(5, 1, 1)

plt.figtext(0.5, 0.300, 'Conclusión Gráfico 1: No parece haber una correlación clara entre la velocidad relativa y la distancia mínima de aproximación. Los asteroides pueden acercarse a diferentes distancias con velocidades variadas.', wrap=True, horizontalalignment='center', fontsize=12) 
plt.figtext(0.5, 0.200, 'Conclusión Gráfico 2: La mayoría de los asteroides en los datos tienen un diámetro máximo estimado de menos de 1 kilómetro, con una caída pronunciada en frecuencia a medida que el diámetro aumenta. Esto sugiere que los asteroides más grandes son menos comunes.', wrap=True, horizontalalignment='center', fontsize=12)
plt.figtext(0.5, 0.100, 'Conclusión Gráfico 3: Hay una variabilidad considerable en el número de aproximaciones cercanas en los años 1900, 1973 y 2106. Esto podría deberse a la variabilidad natural en las órbitas de los asteroides y la detección de estos eventos.', wrap=True, horizontalalignment='center', fontsize=12)

plt.tight_layout()
plt.show()
```
